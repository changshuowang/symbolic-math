```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
train_loss = [2.6466, 2.6354, 2.6187, 2.6124, 2.6627, 2.6050, 2.5938, 2.6720, 2.6205, 2.5954, 2.5843, 2.0370]
train_accuracy =[0.2004, 0.2041, 0.2087, 0.2104, 0.1974, 0.2117, 0.2144, 0.1927, 0.2083, 0.2145, 0.2164, 0.3337] 
test_loss = [3.3857, 3.0366, 3.7088, 3.3726, 3.1248, 3.2686, 4.4613, 3.0416, 4.0395, 4.2540, 4.9851, 2.3441]
test_accuracy =[0.1322, 0.1322, 0.1322, 0.1322, 0.1322, 0.1322, 0.1322, 0.1322, 0.1322, 0.1322, 0.1322, 0.2259] 
```
